```python
import pandas as pd

df = pd.read_csv("Matches.csv")

df.head()
```

```python
df.columns
```

```python
df.info()
```

```python
# DATA CLEANING


# Convertir fecha
df["MatchDate"] = pd.to_datetime(df["MatchDate"])

# Eliminar duplicados
df = df.drop_duplicates()

# Eliminar nulos importantes
df = df.dropna(subset=["FTHome", "FTAway"])

# Crear columna TotalGoals
df["TotalGoals"] = df["FTHome"] + df["FTAway"]

# Eliminar columnas innecesarias
df = df.drop(columns=["HomeShots", "AwayShots"])

# Resetear índice
df = df.reset_index(drop=True)
```

```python
df.head()
```

```python
columns_to_keep = [
    "Division",
    "MatchDate",
    "HomeTeam",
    "AwayTeam",
    "FTHome",
    "FTAway",
    "FTResult",
    "TotalGoals"
]

df = df[columns_to_keep]
```

```python
df.head()
```

```python
df.head()
```

-Los equipos que juegan como locales ganan un mayor porcentaje de partidos que los equipos visitantes.

```python
home_wins = df[df["FTResult"] == "H"]
away_wins = df[df["FTResult"] == "A"]
draws = df[df["FTResult"] == "D"]

total = len(df)

print("Home wins %:", len(home_wins)/total*100)
print("Away wins %:", len(away_wins)/total*100)
print("Draws %:", len(draws)/total*100)
```

-Existen equipos que destacan por anotar una mayor cantidad total de goles a lo largo del tiempo.

```python
home_goals = df.groupby("HomeTeam")["FTHome"].sum()
away_goals = df.groupby("AwayTeam")["FTAway"].sum()

total_goals = home_goals.add(away_goals, fill_value=0).astype(int)

total_goals.sort_values(ascending=False).head(10)
```

-Algunas ligas presentan un mayor promedio de goles por partido en comparación con otras.

```python
goals_by_league = df.groupby("Division")["TotalGoals"].mean()

goals_by_league.sort_values(ascending=False).head(10)
```

```python
league_names = {
    "E0": "Premier League (England)",
    "SP1": "La Liga (Spain)",
    "D1": "Bundesliga (Germany)",
    "F1": "Ligue 1 (France)",
    "I1": "Serie A (Italy)",
    "N1": "Eredivisie (Netherlands)",
    "NOR": "Norwegian League",
    "SUI": "Swiss League",
    "SC2": "Scottish Championship",
    "SC3": "Scottish League One",
    "B1": "Belgian League",
    "AUT": "Austrian League",
    "CHN": "Chinese Super League",
    "USA": "MLS (USA)"
}
```

```python
goals_by_league = goals_by_league.rename(index=league_names)
```

```python
goals_by_league.sort_values(ascending=False).head(10)
```

-La proporción de victorias, empates y derrotas varía dependiendo de la liga.

```python
results_by_league = df.groupby("Division")["FTResult"].value_counts(normalize=True)

results_by_league
```

```python
url = "https://api.football-data.org/v4/competitions"

response = requests.get(url, headers=headers)
api_competitions = response.json()

api_competitions
```

```python
competitions_list = []

for comp in api_competitions["competitions"]:
    competitions_list.append({
        "Competition": comp["name"],
        "Country": comp["area"]["name"],
        "Code": comp["code"]
    })

competitions_df = pd.DataFrame(competitions_list)

competitions_df
```

```python
url = "https://api.football-data.org/v4/competitions/PL/teams"

response = requests.get(url, headers=headers)
api_teams = response.json()

api_teams
```

```python
teams_list = []

for team in api_teams["teams"]:
    teams_list.append({
        "Team": team["name"],
        "Country": team["area"]["name"],
        "Founded": team.get("founded"),
        "Venue": team.get("venue")
    })

teams_df = pd.DataFrame(teams_list)

teams_df
```

```python
print("Número de equipos en Premier League:", len(teams_df))
```

```python
url = "https://api.football-data.org/v4/competitions/PL/standings"

response = requests.get(url, headers=headers)
standings_data = response.json()

standings_data
```

```python
table = standings_data["standings"][0]["table"]

standings_list = []

for team in table:
    standings_list.append({
        "Position": team["position"],
        "Team": team["team"]["name"],
        "Points": team["points"],
        "Played": team["playedGames"],
        "Won": team["won"],
        "Draw": team["draw"],
        "Lost": team["lost"]
    })

standings_df = pd.DataFrame(standings_list)

standings_df
```
